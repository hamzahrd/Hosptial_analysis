```python

def custom_date_parser(date_string):
    return pd.to_datetime(date_string.replace('|', '-'), format='%Y-%m-%dT%H:%M:%SZ')

df2['START'] = df2['START'].apply(custom_date_parser)

print(df2['START'].dtype)
```

```python
df2['STOP'] = pd.to_datetime(df2['STOP'], format='%Y-%m-%dT%H:%M:%SZ')

print(df2['STOP'].dtype)
```

```python
df2['TOTAL_CLAIM_COST'] = df2['TOTAL_CLAIM_COST'].str.replace(',', '.', regex=False)
```

```python
df2['TOTAL_CLAIM_COST'] = pd.to_numeric(df2['TOTAL_CLAIM_COST'])
df2['TOTAL_CLAIM_COST'].dtypes
```
